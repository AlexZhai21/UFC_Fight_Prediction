```python
def advance_letter_string_module(char, step=1):
    if not char.isalpha() or len(char) != 1:
        raise ValueError("Input must be a single alphabet character.")

    if char.islower():
        alphabet = string.ascii_lowercase
    else:
        alphabet = string.ascii_uppercase

    current_index = alphabet.index(char)
    new_index = (current_index + step) % len(alphabet) # Handle wrapping around 'z' or 'Z'
    return alphabet[new_index]
```

```python
letter = "a"
df2_created = False
unique = set()
for i in range(26):
    base_url = "http://ufcstats.com/statistics/fighters?char=" + letter + "&page=all"
    page = ''
    while page == '':
        try:
            page = requests.get(base_url)
            break
        except requests.exceptions.ConnectionError:
            print("Connection refused by the server..")
            print("Let me sleep for 5 seconds")
            print("ZZzzzz...")
            time.sleep(5)
            print("Was a nice sleep")
    soup = BeautifulSoup(page.text, "html.parser")
    body = soup.find_all("tbody")[0]
    extra_links = body.find_all("a")
    href_links = []
    for link in extra_links:
        href_links.append(link.get("href"))
        
    if not df2_created:
        first_fighter = requests.get(href_links[0])
        first_soup = BeautifulSoup(first_fighter.text, "html.parser")
        titles2 = first_soup.find_all("i", class_ = "b-list__box-item-title")
        titles2_test = [title.get_text(strip=True) for title in titles2 if title.text.strip()]
        good_titles = ["Name"] + titles2_test[6:]
        df2 = pd.DataFrame(columns=good_titles)
        df2_created = True
    for fighter in href_links:
        if fighter in unique:
            continue
        else:
            unique.add(fighter)  
        fighter_page = requests.get(fighter)
        fighter_soup = BeautifulSoup(fighter_page.text, "html.parser")
        fighter_row_data = fighter_soup.find_all("li", class_ = "b-list__box-list-item b-list__box-list-item_type_block")
        messy_row = [row.text.strip() for row in fighter_row_data if row.text.strip()]
        cleaned_row = [value.split("\n")[-1].strip() for value in messy_row if value.strip()]
        actual_row = cleaned_row[-8:]
        name = fighter_soup.find_all("h2", class_ = "b-content__title")[0].text.strip().split("\n")[0]
        row = [name] + actual_row
        df2length = len(df2)
        df2.loc[df2length] = row
    print("Just finished " + letter)
    letter = advance_letter_string_module(letter)
```
