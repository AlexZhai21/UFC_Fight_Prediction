```python
#Binning Current Win Streak Difference:
def win_streak_bins(x):
    if x == 0:
        return 0 #Same win streak
    elif   1 <= x <= 2:
        return 1 #Fighter win streak > Opponent by a little bit
    elif 2 < x:
        return 2 #
    elif -2 <= x <= -1:
        return -1
    else:
        return -2
df["Current_Win_Streak_Bins"] = df["Current_Win_Streak_Diff"].apply(win_streak_bins)
```

```python
#Binning wins by ko_%:
def win_by_ko_perc(x):
    if x == 0:
        return 0 #Same win % by KO
    elif 0 < x < 0.25:
        return 1 #Fighter win% by KO slightly above opponent
    elif x > 0.25:
        return 2 #Fighter win% by KO greatly above opponent
    elif -0.25 < x < 0:
        return -1 #Fighter win% by KO slightly below opponent
    else:
        return -2 #Fighter win% by KO greatly below opponent
df["Wins_By_KO_%_Bins"] = df["Wins_By_KO_%_Diff"].apply(win_by_ko_perc)
```

```python
fig, ax = plt.subplots(2, figsize=(15, 10))

sns.histplot(data = df, x = "Wins_By_KO_%_Bins", hue="Is_Winner_x", ax = ax[0])
sns.boxplot(data = df, x = "Wins_By_KO_%_Bins", hue="Is_Winner_x", ax = ax[1]) #Nearly identical so probably not a useful feature at all
```

```python
def box_and_hist(data, x, hue):
    fig, ax = plt.subplots(2, figsize=(15,10))
    sns.histplot(data = data, x = x, hue = hue, ax = ax[0])
    sns.boxplot(data=data, x = x, hue = hue, ax = ax[1]) #created a function to quicly plot both a boxplot and a histplot.
```
