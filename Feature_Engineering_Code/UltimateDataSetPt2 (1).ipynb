```python
def win_loss_methods(df, method, fighter, won):
    #method is KO, SUB or Dec, fighter is either fighter or opponent
    if won:
    
        if fighter == "Fighter":
            bool_val = True
        else:
            bool_val = False
            
        df[fighter + "_"+ method + "_WIN"] = ((df["Method"] == method) & (df["Is_Winner"] == bool_val)).astype(int)
     
        df[fighter + "_Wins_By_" + method] = df.groupby(fighter)[fighter + "_"+ method + "_WIN"].transform(lambda x: x.cumsum().shift(1).fillna(0))
        df.drop(fighter + "_"+ method + "_WIN", axis=1, inplace=True)
        return df, fighter + "_Wins_By_" + method
    elif not won:
        if fighter == "Fighter":
            bool_val = False
        else:
            bool_val = True
            
        df[fighter + "_"+ method + "_LOST"] = ((df["Method"] == method) & (df["Is_Winner"] == bool_val)).astype(int)
      
        df[fighter +  "_Loss_By_" + method] = df.groupby(fighter)[fighter + "_"+ method + "_LOST"].transform(lambda x: x.cumsum().shift(1).fillna(0))
        df.drop(fighter + "_"+ method + "_LOST", axis=1, inplace=True)
        return df, fighter +  "_Loss_By_" + method


        
    
    
```

```python
names = []

for fighter in ["Fighter", "Opponent"]:
    for won in [True, False]:
        for method in ["KO", "SUB", "DEC"]:

   
            names.append(win_loss_methods(methods, method, fighter, won)[1])
    
```

```python
#Temporary: Reverse Won_last_fight and UFC_Debut
final.loc[final["UFC_Debut"] == True, "Won_Last_Fight"] = False
final.loc[final["Opp_UFC_Debut"] == True, "Opp_Won_Last_Fight"] = False
```

```python
def compute_win_streak(wins):
    streak = 0
    streaks = []
    for result in wins:
        if result:
            streak += 1
        else:
            streak = 0
        streaks.append(streak)
    return streaks

final["Fighter_Current_Win_Streak"] = (final.groupby("Fighter")["Won_Last_Fight"].transform(compute_win_streak))
```

```python
final["Opp_Current_Win_Streak"] = (final.groupby("Opponent")["Opp_Won_Last_Fight"].transform(compute_win_streak))
```
